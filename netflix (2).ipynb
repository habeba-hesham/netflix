```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='type', palette='Purples')
plt.title('Count of Movies and TV Shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,5))
top_countries = df['country'].value_counts().head(5)
sns.barplot(x=top_countries.index, y=top_countries.values, palette='magma')
plt.xticks(rotation=45)
plt.title('Top 10 Countries with Most Netflix Content')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.countplot(data=df, y='rating',order=df['rating'].value_counts().index[:10], palette='pink')
plt.title('Most Common Ratings on Netflix')
plt.xlabel('Count')
plt.ylabel('Rating')
plt.show()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], format='%B %d, %Y', errors='coerce')
```

```python
df['month_added'] = df['date_added'].dt.month_name()
plt.figure(figsize=(12, 6))
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']
sns.countplot(data=df, x='month_added', order=month_order, palette='rocket')
plt.title('Content Added by Month')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
import pandas as pd
import numpy as np
from PIL import Image

# لنفترض إن عندك DataFrame اسمه df
# والعمود اللي فيه النصوص اسمه 'اسم_العمود'

# دمج كل النصوص من العمود في نص واحد كبير
text = ' '.join(df['month_added'].astype(str))

# تحديد الكلمات اللي مش عايزين تظهر في السحابة (كلمات إيقاف)
stopwords = set(STOPWORDS)
# ممكن تضيف كلمات إيقاف مخصصة لو عايز
# stopwords.add("كلمة_مش_عايزها")

# تحديد شكل القناع (اختياري)
mask = None
# لو عايز تستخدم قناع من صورة:
# mask = np.array(Image.open("اسم_الصورة.png"))

# إنشاء كائن WordCloud
wc = WordCloud(background_color="white",
               max_words=200,
               stopwords=stopwords,
               mask=mask,
               contour_width=3,
               contour_color='steelblue')

# توليد سحابة الكلمات من النص المدمج
wc.generate(text)

# عرض سحابة الكلمات باستخدام Matplotlib
plt.figure(figsize=(12, 12))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()

# لحفظ الصورة (اختياري)
# wc.to_file("wordcloud_from_column.png")
```
